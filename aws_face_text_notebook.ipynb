# üß† Face & Text Extraction com AWS

Este notebook mostra como usar os servi√ßos Rekognition e Textract da AWS para:
- Detectar rostos em imagens
- Extrair textos (OCR)

```
import boto3
import os
from dotenv import load_dotenv
from PIL import Image
import io

# Carrega vari√°veis da AWS do .env
load_dotenv()
rekognition = boto3.client(
    "rekognition",
    aws_access_key_id=os.getenv("AWS_ACCESS_KEY_ID"),
    aws_secret_access_key=os.getenv("AWS_SECRET_ACCESS_KEY"),
    region_name=os.getenv("AWS_REGION")
)

textract = boto3.client(
    "textract",
    aws_access_key_id=os.getenv("AWS_ACCESS_KEY_ID"),
    aws_secret_access_key=os.getenv("AWS_SECRET_ACCESS_KEY"),
    region_name=os.getenv("AWS_REGION")
)
```

## üì§ Enviar imagem para an√°lise

```
# Caminho para a imagem local
image_path = "sua_imagem.jpg"

# Carregar e converter para bytes
with open(image_path, "rb") as f:
    image_bytes = f.read()
```

## üßë‚Äçü¶± Detectar rostos com Rekognition

```
response_faces = rekognition.detect_faces(
    Image={"Bytes": image_bytes},
    Attributes=["ALL"]
)
for i, face in enumerate(response_faces["FaceDetails"]):
    print(f"Rosto {i+1} - Confian√ßa: {face['Confidence']:.2f}%")
```

## üìù Extrair texto com Textract

```
response_text = textract.detect_document_text(
    Document={"Bytes": image_bytes}
)

textos = [block["Text"] for block in response_text["Blocks"] if block["BlockType"] == "LINE"]
print("\n".join(textos))
```
